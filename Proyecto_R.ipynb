## Accidentes de Tránsito en los Estados Unidos

<img src="https://st1.uvnimg.com/dims4/default/3ae48bc/2147483647/thumbnail/1024x576%3E/quality/75/?url=http%3A%2F%2Fuvn-brightspot.s3.amazonaws.com%2F74%2Ff0%2F7b17341b44378d57d379833ba5a8%2Fgettyimages-670437722.jpg" width = 500>

Las muertes por automóviles son una imperiosa preocupación de salud pública. Se trata de **la segunda causa de muerte accidental en el país**, después de las sobredosis por drogas. Sin embargo, las legislaturas estatales están rezagadas de cara a esta realidad y no toman las medidas necesarias para desacelerar el ritmo de muertes y accidentes, según informa un grupo defensor de la seguridad vial.

En el informe de 2018 relativo a las leyes de tránsito estatales, la organización *Advocates for Highway and Auto Safety* (Defendiendo la Seguridad Vial y Automotriz) concluye que solo seis estados muestran “avances significativos” en pos de adoptar 16 leyes que ellos consideran “esenciales para salvar vidas, evitar lesiones y reducir cuidados médicos y otros costos”.

Por ejemplo: **cerca de la mitad del total de automovilistas que murieron en accidentes durante los últimos 5 años no tenían puesto el cinturón de seguridad**, mientras que 16 estados todavía carecen de regulaciones para reforzar el uso obligatorio de éste. Asimismo, la tasa de muertes por conducir bajo los efectos del alcohol está aumentando; Pese a ello, **20 estados no han aprobado normativas que requieran alcoholímetros para previos infractores**, y 9 no disponen de leyes para fiscalizar la presencia de una lata o una botella abierta en el auto. Y aunque el 40% de los motociclistas fallecidos no llevaba casco, más estados rechazan los requerimientos de estos medios de protección que los que los imponen.

Pero algo mueve al optimismo: 13 leyes de “seguridad básica” fueron aprobadas estatalmente en 2017, según el reporte. En 2015 y 2016 solo habían ocho y cinco leyes de ese tipo, respectivamente.

Por otra parte, las leyes son tan buenas como se hacen cumplir. Un buen ejemplo: los choferes que tienen sus ojos, manos, y cerebros pendientes al sonido del teléfono o a los sistemas de info-entretenimiento y navegación de sus vehículos están complicando el panorama. Las así llamadas muertes por distracción al volante han crecido en muchos de los estados que aprobaron regulaciones para controlar el número de estas, ya fuera a través de dispositivos manuales o prohibiciones de mensajes de texto. Eso se ha debido en gran medida a que, según el informe, **el modo de hacer cumplir estas leyes ha sido laxo y fácil de evadir**. Más aún, pese a que la distracción con los llamados manos libres **podría ser tan peligrosa como con las manos ocupadas**, ningún estado ha vetado el uso de celulares en general.

Por si no bastara, las condiciones en carretera podrían ser incluso más peligrosas de lo que sugiere el reporte, ya que éste no evalúa las leyes de límite de velocidad. El número de estados subiendo sus límites ha aumentado recientemente, siendo Nevada el último en permitir que los choferes lleguen a 80 millas por hora en algunos tramos de las autopistas. La velocidad, ni qué decir tiene, es una de las causas de accidentes mortales más ampliamente reportadas, y **podría ser uno de los factores del que más al tanto estén los legisladores** según la **Junta Nacional de Seguridad del Transporte** (NTSB, por sus siglas en inglés). **Bajar los límites de velocidad pudiera salvar vidas**, evidencia su investigación; Francia, recientemente, rebajó esos límites en las autopistas de doble sentido, en respuesta al creciente número de muertes en carretera.

La NTSB instó a todos los estados a que revisaran, radicalmente, **cómo establecen sus límites de velocidad**: no solo hallando la velocidad promedio de los automovilistas, como invitó a hacer por mucho tiempo un desactualizado estudio federal, sino “también teniendo en cuenta las condiciones de la carretera, el estado de la infraestructura al costado de las vías, el estacionamiento, la presencia de peatones y el historial de accidentes en la zona”, precisó la revista *Governing*. Estas recomendaciones son muy pertinentes, habida cuenta que los estados rara vez consideran las condiciones locales al establecer sus límites de velocidad.

Menos y más estrechos carriles, y la protección de los cruces para peatones y ciclistas –cuyo índice de fatalidades en carreteras subieron, respectivamente, un 9% y un 1.3% en 2016– han demostrado que salvan vidas haciendo reducir la velocidad y amparando a los usuarios más vulnerables. Decenas de ciudades, un puñado de estados, e incluso el Departamento de Transporte de Estados Unidos han adoptado o modelado políticas tras Vision Zero, la campaña internacional encaminada a poner fin a los accidentes automovilísticos nocivos por medio de una mejor vigilancia de las leyes, de educación para la seguridad y una ingeniería civil intencionada. Los mejores resultados hasta ahora parecen estar relacionados con ese última “I” –aunque las muertes de peatones y ciclistas en la ciudad de Nueva York, en general, **han aumentado** desde que la urbe se comprometió seguir la pauta de Vision Zero, las muertes **en intersecciones especialmente rediseñadas han disminuido**.

Según la Organización Mundial de la Salud, **Estados Unidos se ubica en la posición 41 entre las 52 naciones de más altos ingresos en lo relacionado a muertes por accidentes de tránsito**, superado solo por Qatar, Arabia Saudita y Rusia.

### Las muertes por accidentes de tránsito en Estados Unidos Aumentan incluso cuando la pandemia reduce los kilómetros recorridos

<img src="https://www.wfla.com/wp-content/uploads/sites/71/2021/03/Crashes-spike-despite-pandemic.jpg?w=960&h=540&crop=1" width = 600 >

Los cierres pandémicos y las órdenes de quedarse en casa mantuvieron a muchos conductores alejados de las carreteras y autopistas estadounidenses el año pasado. Pero aquellos que se aventuraron encontraron carriles abiertos que solo invitaban a conducir imprudentemente, lo que provocó un fuerte aumento en las muertes por accidentes de tránsito en todo el país.

El Consejo Nacional de Seguridad, una organización sin fines de lucro, estima en un informe publicado el jueves que 42,060 personas murieron en choques de vehículos en 2020, un aumento del 8% con respecto a 2019 y el primer salto en cuatro años.

Además, la tasa de mortalidad por cada 100 millones de millas conducidas aumentó un 24%, el mayor aumento porcentual anual desde que el consejo comenzó a recopilar datos en 1923.

Y aunque el tráfico ahora se está acercando a los niveles previos al coronavirus, el mal comportamiento en las carreteras continúa, dicen las autoridades.

“Es algo aterrador lo que estamos viendo en nuestras carreteras”, dijo Michael Hanson, director de la Oficina de Seguridad Vial del Departamento de Seguridad Pública de Minnesota. “Estamos viendo un gran aumento en la cantidad de comportamientos de riesgo”.

Las muertes del año pasado fueron la mayor cantidad desde 2007, cuando 43,945 personas murieron en choques de vehículos. Además, el consejo de seguridad estima que 4,8 millones de personas resultaron heridas en accidentes el año pasado.

Los datos federales muestran que los estadounidenses condujeron un 13% menos de millas el año pasado, o aproximadamente 2.8 billones de millas, dijo Ken Kolosh, gerente de estadísticas del consejo de seguridad. Sin embargo, el número de muertes aumentó a un ritmo alarmante, dijo.

“La pandemia parece estar quitando nuestros ojos de la pelota cuando se trata de seguridad vial”, dijo Kolosh.

De los comportamientos imprudentes, los primeros datos de la Administración Nacional de Seguridad del Tráfico en las Carreteras muestran que la velocidad es el factor principal, dijo Kolosh. Además, las pruebas de los pacientes del centro de trauma involucrados en accidentes de tránsito muestran un mayor uso de alcohol, marihuana y opiáceos, dijo.

En Minnesota, los volúmenes de tráfico cayeron un 60% cuando se emitieron órdenes para quedarse en casa a principios de la pandemia la primavera pasada. Hanson dijo que los funcionarios estatales esperaban una caída correspondiente en choques y muertes, pero mientras que los choques disminuyeron, las muertes aumentaron.

“Casi de inmediato, la tasa de mortalidad comenzó a aumentar, y a aumentar significativamente”, dijo Hanson, y agregó que sus contrapartes en otros estados vieron aumentos similares. “Creó menos congestión y mucho más espacio en los carriles para que los buceadores lo usaran y, honestamente, para abusar”.

A fines de marzo y principios de abril, la cantidad de muertes relacionadas con la velocidad se duplicó con creces durante el mismo período en 2019 en el estado, dijo Hanson. El año pasado, Minnesota registró 395 muertes por accidentes de tráfico, casi un 9% más que las 364 de 2019.

Los conductores también utilizaron las carreteras vacías para conducir a velocidades extremas. En 2019, los 600 soldados de la Patrulla Estatal de Minnesota entregaron multas a poco más de 500 conductores por superar las 100 mph (160 kph). Ese número aumentó a 1.068 en 2020, dijo Hanson.

Viajar a más de 100 mph hace que los choques sean mucho más severos, dijo el consejo de seguridad.

El alto número de conductores a exceso de velocidad continúa incluso cuando el tráfico está comenzando a regresar a niveles previos a la pandemia, según Hanson.

El consejo de seguridad pide una aplicación equitativa de las leyes de tránsito, mejoras de infraestructura, cerraduras de encendido obligatorias para conductores ebrios condenados, reducción de los límites de velocidad para que coincidan con los diseños de las carreteras y leyes que prohíben el uso de teléfonos celulares mientras se conduce, entre otras recomendaciones para detener las muertes.

El consejo recopila datos de accidentes fatales de los estados en carreteras públicas y privadas. Las cifras publicadas el jueves son preliminares, pero cada año es solo ligeramente diferente de las cifras finales, dijo Kolosh.

#### Clima en los Estados Unidos

Los Estados Unidos contiguos cubren un área enorme y, por lo tanto, es lógico que presenten una notable variedad climática. En general, sin embargo, tienen un clima continental, con inviernos fríos (a menudo gélidos) y veranos calurosos (a menudo muy calurosos), con diferentes estaciones dependiendo de la latitud y de la distancia del mar. Sin embargo, hay excepciones: la costa del Pacífico occidental, fresca y húmeda en el norte y mediterránea en el sur; la costa del golfo de México, templada en invierno y bochornosa en verano; Florida, casi tropical; las zonas montañosas, frías todo el año; el área de los desiertos cálidos.
Como no hay obstáculos para el descenso del aire frío desde Canadá, casi todo el país puede experimentar repentinas olas frío en invierno, que sin embargo tienen una intensidad y duración diferentes según el área. Los períodos fríos duran unos días en el sur, donde la temperatura desciende unos pocos grados bajo cero en invierno, mientras que son intensos y a veces largos en las zonas del interior, en las mesetas y en el noreste. Incluso las olas de calor del verano pueden ser intensas, especialmente en las zonas del interior. En general, la mitad occidental es más árida que la mitad oriental, con la excepción de la costa del Pacífico centro-norte, que es lluviosa.

**¿ Como influyen las condiciones climáticas en la provocación de accidentes de viales ?**

```r
library(ggcorrplot)
acc <- accidentes %>%
        select(Severity,Temperature:Precipitation)
corr <- round(cor(acc),1)
ggcorrplot(corr, hc.order=TRUE, type="lower", lab=TRUE, lab_size=3, method= "square",
          colors=c("red","white","green"))
```

Realmente las variables climáticas tienen una correlación baja entre ellas como para agruparlas, por lo que para analizar la influencia de las mismas en accidentes de tránsito se hará de forma individual.

**¿ Cúal es el número de accidentes de tránsito por condiciones del tiempo ?**

```r
accidentes %>%
        group_by(Weather_Condition) %>%
        summarise(Total = n()) %>%
        arrange(Total) %>%
        tail(n = 10) %>%
        ggplot(aes(x=Weather_Condition, y=Total)) +
        geom_bar(stat="identity", width=.5,fill="darkviolet", color="black") +
        labs(title="Accidentes de Tránsito por Condiciones del Tiempo", subtitle="Año 2020", 
             x="Condiciones del Tiempo", y="Número de Accidentes") +
        theme_bw()
```

En un sentido mas específico esta palabra puede ser usada en un pronóstico u observación del servicio meteorológico, cuando menos de 3/8 del cielo está cubierta por nubes opacas, no hay precipitación, la visibilidad es ilimitada y no hay extremos de temperatura o vientos (lo que generalmente significa que no es anormalmente frío o cálido para la temporada y la hora del día y que los vientos son unos 10 nudos o menos.

**¿ Cuál es la hora promedio donde mas accidentes se originan ?**

```r
accidentes %>%
    mutate(Start_Time =ymd_hms(Start_Time) ) %>%
    mutate(Hour = hour(Start_Time)) %>%
    group_by(Hour) %>%
    summarise(Total = n()) %>%
    ggplot(aes(Hour,Total)) +
    geom_point(col="darkorange",size=4) +
    labs(title="Cantidad Acumulada de Accidentes de Tránsito Reportados en cada Hora", subtitle="Año 2020",
          x="Horas del Día", y="Total de Accidentes") +
    theme_bw()
```

La hora en que se reportaron mayor cantidad de accidentes durante el año 2020 fue apróximadamente las **4 de la tarde**

**¿ En qué lapso de la semana se originan más accidentes ?**

```r
acc_week <- accidentes %>%
            mutate(Start_Time = ymd_hms(Start_Time), hora = hour(Start_Time)) %>%
            mutate(dia_semana = wday(Start_Time), finde = ifelse(dia_semana == 6 | dia_semana == 7, "Weekend", "Weekday" )) %>%
            group_by(finde, hora) %>%
            summarise(Total = n(), .groups = 'drop')%>%
            spread(finde, Total)  
```

```r
g1 <-ggplot(acc_week,aes(x=hora,y=Weekday)) +
        geom_point(col="orange",size=4) +
        labs(title = "Promedio de Accidentes de Tránsito por Cada Hora", subtitle="Lunes a Viernes/Año 2020",
            y="Número de accidentes", x="Horas") +
        theme_bw()

g2 <- ggplot(acc_week, aes(x=hora,y=Weekend)) +
        geom_point(col="green3",size=4) +
        labs(title = "Promedio de Accidentes de Tránsito por Cada Hora", subtitle = "Fines de Semana/Año 2020",
            y= "Número de Accidentes", x ="Horas") +
        theme_bw()

grid.arrange(g1,g2,ncol=2)
```

El comportamiento de accidentes de tránsito durante los fines de semana en comparación con los otros días de la semana es muy similar, mas en cantidad de accidentes la diferencia es bastante amplia. En ambos casos las horas de menor cantidad de incidentes son entre las **12 de la noche** y **5 de la mañana**. Mientras que entre las **5 y 10 de la mañana** tiende a incrementarse mas las incidencias durante los días de entre semana y la hora de mayor cantidad de accidentes en ambos casos son las **4 de la tarde**.

**¿ En que mes son mas comúnes los accidentes de tránsito ?**

```r
accidentes %>%
    mutate(Start_Time = ymd_hms(Start_Time), mes = month(Start_Time, label=TRUE)) %>%
    group_by(mes) %>%
    summarise(Total = n()) %>%
    ggplot(aes(x=mes,y=Total)) +
    geom_bar(stat="Identity", width = .8, fill="cyan4",color="black") +
    labs(title="Total de Accidentes de Tránsito por Mes",subtitle="Año 2020",x="Meses",
         y="Cantidad de Accidentes") +
    theme_bw()
```

El mes de **diciembre** es el que presentó mayor cantidad de accidentes de tránsito durante el 2020. Además hubo una tendencia de crecimiento durante los últimos tres meses del año. Por el contrario **julio y agosto** tuvieron una cantidad baja de accidentes.

**¿ Qué día de la semana es donde se reportaron más accidentes de tránsito ?**

```r
accidents_dia <- accidentes %>%
                    mutate(Start_Time = ymd_hms(Start_Time),
                           Dia = wday(Start_Time, label = TRUE, abbr=FALSE)) %>%
                    group_by(Dia) %>%
                    summarise(Total = n())
```

```r
#Se genera el gráfico de accidentes por día
#Se almacena en un variable para gráficarlo en grid 
g1 <- ggplot(accidents_dia,aes(x=Dia,y=Total)) +
        geom_bar(stat = "Identity", width = 0.6, fill="orange",color="black") +
        labs(title="Número de Accidentes Acumulado, Reportados por Día", subtitle="Año 2020",
            y = "Número de Accidentes", x="Días") +
         theme_bw()
```

```r
#Se genera el gráfico de distribución de accidentes por día
distribucion <- accidentes %>%
                    mutate(Start_Time = ymd_hms(Start_Time),
                          Dia = wday(Start_Time))
#Se almacena en una variable para graficarlo en un grid
g2 <- ggplot(distribucion,aes(Dia)) +
         geom_histogram(aes(fill=State),binwidth = .5,col="white",size=1) +
         labs(title = "Distribución de Accidentes por Día", subtitle="Año 2020",y="Número de Accidentes") +
         theme_bw()
```

```r
grid.arrange(g1,g2,ncol=2)
```

El día **jueves** es cuando ocurren mas accidentes de tránsito y el **domingo** cuando hay menos incidencias.
